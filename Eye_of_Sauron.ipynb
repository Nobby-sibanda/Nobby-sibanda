```python
#METHOD 1:IP Address geolocation when connected to the wifi (given many ip addresses)

import requests

def get_location_from_ip(ip_address=None, person_name="Unknown"):
    """Get location from IP address with person identification"""
    try:
        if ip_address:
            response = requests.get(f'http://ip-api.com/json/{ip_address}')
        else:
            response = requests.get('http://ip-api.com/json/')

        data = response.json()

        if data['status'] == 'success':
            print(f"\n=== Location for {person_name} ===")
            print(f"IP Address: {data.get('query')}")
            print(f"Country: {data.get('country')}")
            print(f"Region/State: {data.get('regionName')}")
            print(f"City: {data.get('city')}")
            print(f"ZIP Code: {data.get('zip')}")
            print(f"Coordinates: ({data.get('lat')}, {data.get('lon')})")
            print(f"ISP/Carrier: {data.get('isp')}")
            print(f"Timezone: {data.get('timezone')}")
            print(f"Organization: {data.get('org')}")

            return {
                'name': person_name,
                'ip': data.get('query'),
                'city': data.get('city'),
                'region': data.get('regionName'),
                'country': data.get('country'),
                'zip': data.get('zip'),
                'latitude': data.get('lat'),
                'longitude': data.get('lon'),
                'isp': data.get('isp'),
                'timezone': data.get('timezone')
            }
        else:
            print(f"Failed to get location for {person_name}: {data.get('message')}")
            return None

    except Exception as e:
        print(f"Error getting location for {person_name}: {e}")
        return None
    except requests.exceptions.RequestException as e:
      print(f"Network error: {e}")
    except KeyError as e:
      print(f"Missing data field: {e}")
    except Exception as e:
      print(f"Unexpected error: {e}")

# Track multiple IP addresses
Friends = {
    "Nobby": "147.182.172.178",
    "Fungai": "3.101.29.241",
    "Hagai": "47.89.192.121",
    "Andiswa": "155.2.215.150"
}

# Collect all locations
all_locations = []

for name, ip in Friends.items():
    location = get_location_from_ip(ip, name)
    if location:
        all_locations.append(location)

# Optional: Save to CSV
import csv

with open('phone_locations.csv', 'w', newline='') as f:
    if all_locations:
        writer = csv.DictWriter(f, fieldnames=all_locations[0].keys())
        writer.writeheader()
        writer.writerows(all_locations)
        print("\n✅ Data saved to phone_locations.csv")
```

```python
#Method 2: GPS Coordinates (from phone's GPS)
# From photos (EXIF data)
!{sys.executable} -m pip install pillow exifread

from PIL import Image
from PIL.ExifTags import TAGS, GPSTAGS

def get_gps_from_image(image_path):
    """Extract GPS coordinates from image metadata"""
    image = Image.open(image_path)
    exif_data = image._getexif()

    if not exif_data:
        print("No EXIF data found")
        return None

    for tag_id, value in exif_data.items():
        tag = TAGS.get(tag_id, tag_id)
        if tag == "GPSInfo":
            gps_data = {}
            for key in value.keys():
                decode = GPSTAGS.get(key, key)
                gps_data[decode] = value[key]

            # Convert to decimal coordinates
            lat = gps_data.get('GPSLatitude')
            lon = gps_data.get('GPSLongitude')

            if lat and lon:
                print(f"Latitude: {lat}")
                print(f"Longitude: {lon}")
                return gps_data

    return None

# Example: get_gps_from_image('photo.jpg')
```

```python
# Method 3:  Cell Tower Triangulation
#Using OpenCelliD API (requires cell tower data)
!{sys.executable} -m pip install requests

def get_location_from_cell_tower(mcc, mnc, lac, cell_id, api_key):
    """
    Get location from cell tower information
    mcc: Mobile Country Code
    mnc: Mobile Network Code
    lac: Location Area Code
    cell_id: Cell Tower ID
    """
    url = f"https://opencellid.org/cell/get?key={api_key}&mcc={mcc}&mnc={mnc}&lac={lac}&cellid={cell_id}&format=json"

    response = requests.get(url)
    data = response.json()

    print(f"Latitude: {data.get('lat')}")
    print(f"Longitude: {data.get('lon')}")
    print(f"Accuracy: {data.get('range')} meters")

    return data
```

```python
#Method 4: WiFi Network Geolocation
#Using Google Geolocation API or similar services
def get_location_from_wifi(bssid_list, api_key):
    """
    Get location from WiFi access points
    bssid_list: List of WiFi MAC addresses the phone can see
    """
    import requests

    url = f"https://www.googleapis.com/geolocation/v1/geolocate?key={api_key}"

    payload = {
        "wifiAccessPoints": [
            {"macAddress": bssid} for bssid in bssid_list
        ]
    }

    response = requests.post(url, json=payload)
    data = response.json()

    print(f"Latitude: {data['location']['lat']}")
    print(f"Longitude: {data['location']['lng']}")
    print(f"Accuracy: {data['accuracy']} meters")

    return data
```

```python
#Method 5:Reverse Phone Lookup Services (Paid/Limited)
#Using services like Twilio Lookup API
!{sys.executable} -m pip install twilio

from twilio.rest import Client

def lookup_phone_details(phone_number, account_sid, auth_token):
    """
    Use Twilio Lookup API (paid service)
    Can provide carrier info and sometimes caller name
    """
    client = Client(account_sid, auth_token)

    number = client.lookups.v2.phone_numbers(phone_number).fetch()

    print(f"Phone Number: {number.phone_number}")
    print(f"Country Code: {number.country_code}")
    print(f"Carrier: {number.carrier}")

    return number
```
