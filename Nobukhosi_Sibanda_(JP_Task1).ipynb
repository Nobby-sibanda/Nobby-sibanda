```python
#to take into account all data points while reducing loss, use a polynomial trend to account for seasonal component

def price_model(days, a, b, c, amp, phase):
    """
    Model: Linear trend + seasonal (annual) component
    days: days since start date
    a, b: linear trend parameters
    c: offset
    amp: amplitude of seasonal variation
    phase: phase shift for seasonal component
    """
    trend = a * days + b * days**2 + c
    seasonal = amp * np.sin(2 * np.pi * days / 365.25 + phase)
    return trend + seasonal
```

```python
def estimate_price(input_date):
    """
    Estimate natural gas price for any given date.

    Parameters:
    input_date: string in format 'MM/DD/YYYY' or datetime object

    Returns:
    float: estimated price
    """

    if isinstance(input_date, str):
        query_date = datetime.strptime(input_date, '%m/%d/%Y')
    else:
        query_date = input_date

    days_from_start = (query_date - start_date).days

    # interpolation
    if days_from_start >= 0 and days_from_start <= days_since_start[-1]:
        interp_func = interp1d(days_since_start, prices, kind='cubic', fill_value='extrapolate')
        return float(interp_func(days_from_start))
    else:
        return float(price_model(days_from_start, *popt))
```

```python
popt, pcov = curve_fit(price_model, days_since_start, prices,
                       p0=[0.001, 0, 10, 0.5, 0], maxfev=10000)
```
